```python
# made this file to create a map key for Toronto neighborhoods we willl use later when we visualize data on a map
import os
import pandas as pd

# Ensure the directory exists
os.makedirs('sets', exist_ok=True)

# Load the 2024 file (it has the most recent neighborhood list)
df_map = pd.read_csv('2024.csv')

# We only need the name and the shape
map_key = df_map[['AREA_NAME', 'geometry_wkt']].drop_duplicates()

# Save it to your new 'processed_data' folder
map_key.to_csv('sets/toronto_map_key.csv', index=False)
print("Map Key Created")
```

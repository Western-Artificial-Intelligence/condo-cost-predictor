```python
# First, let's fix the manual_lease_cols creation and properly add the columns
manual_lease_cols = [
    'Bachelor Leased',
    'bachelor_avg_lease_rate',
    '1_bedrooms_leased',
    '1_bed_room_avg_lease_rate',
    '2_bedrooms_leased',
    '2_bedrooms_avg_lease_rate',
    '3_bedrooms_leased',
    '3_bedrooms_avg_lease_rate'
]

# Add these columns to df_2011 if they don't exist
for col in manual_lease_cols:
    if col not in df_2011.columns:
        df_2011[col] = 0

# Now update the rental columns using the averaged data
rental_cols = manual_lease_cols  # Use the same list

# Remove any existing 'Toronto ' prefix before adding
rental_avg['Area'] = rental_avg['Area'].str.replace('Toronto ', '', regex=False)
rental_avg['Area'] = 'Toronto ' + rental_avg['Area']
rental_avg_indexed = rental_avg.set_index('Area')

print("df_2011 Areas:", df_2011['Area'].unique())
print("rental_avg Areas:", rental_avg['Area'].unique())

# Update rental columns in df_2011 - only for columns that exist
for col in rental_cols:
    if col in df_2011.columns and col in rental_avg_indexed.columns:
        # Count zeros before update
        zero_count = (df_2011[col] == 0).sum()
        print(f"Updating {col}: replacing {zero_count} zeros")
        
        # Perform the update
        df_2011.loc[df_2011[col] == 0, col] = df_2011['Area'].map(rental_avg_indexed[col])
    else:
        print(f"Skipping {col}: not found in both dataframes")

# FINALLY - Save the corrected dataframe to CSV
df_2011.to_csv('../sets/2011_dataframe.csv', index=False)
print("✓ Final 2011 dataset saved to '../sets/2011_dataframe.csv'")
print(f"Final shape: {df_2011.shape}")

# Show the result
df_2011.head()
```

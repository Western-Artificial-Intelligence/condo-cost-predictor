```python
# Check if it matches the required schema
required_cols = ['AREA_NAME', 'CLASSIFICATION', 'CLASSIFICATION_CODE', 'geometry_wkt',
       'geometry_type', 'Area', 'Bachelor Leased', 'bachelor_avg_lease_rate',
       '1_bedrooms_leased', '1_bed_room_avg_lease_rate', '2_bedrooms_leased',
       '2_bedrooms_avg_lease_rate', '3_bedrooms_leased', '3_bedrooms_avg_lease_rate',
       'area_sq_meters', 'perimeter_meters', 'park_count',
       'total_stop_count', 'avg_stop_frequency', 'max_stop_frequency',
       'total_line_length_meters', 'transit_line_density', 'distinct_route_count']

# Add year-specific crime columns
crime_types = ['ASSAULT', 'ASSAULT_RATE', 'AUTOTHEFT', 'AUTOTHEFT_RATE', 
               'BIKETHEFT_RATE', 'BREAKENTER', 'BIKETHEFT_RATE',
               'HOMICIDE', 'HOMICIDE_RATE', 'ROBBERY', 'ROBBERY_RATE',
               'SHOOTING', 'SHOOTING_RATE', 'THEFTFROMMV', 'THEFTFROMMV_RATE',
               'THEFTOVER', 'THEFTOVER_RATE', 'POPULATION']

expected_crime_cols = [f"{crime}_2016" for crime in crime_types]
all_required = required_cols + expected_crime_cols

print(f"Expected columns: {len(all_required)}")
print(f"Current columns: {len(df_2016.columns)}")
print(f"\nRows: {len(df_2016)} (target - 158)")
```

```python
# Check which columns are missing
current_cols = set(df_2016.columns)
required_cols_set = set(all_required)

missing_cols = required_cols_set - current_cols
extra_cols = current_cols - required_cols_set

print(f"Missing columns ({len(missing_cols)}):")
for col in sorted(missing_cols):
    print(f"  - {col}")

print(f"\nExtra columns ({len(extra_cols)}):")
for col in sorted(extra_cols):
    print(f"  - {col}")
```

```python
static_cols = [
    'AREA_NAME',
    'CLASSIFICATION',
    'CLASSIFICATION_CODE',
    'geometry_wkt',
    'geometry_type',
    'Area',
    'area_sq_meters',
    'perimeter_meters',
    'park_count',
    'total_stop_count',
    'avg_stop_frequency',
    'max_stop_frequency',
    'total_line_length_meters',
    'transit_line_density',
    'distinct_route_count',
]

# build df_2010 from df_2024:
# - keep static cols (fill with NA if missing)
# - convert all *_2024 columns into *_2010columns filled with 0
df_2010= pd.DataFrame(index=df_2024.index)

# populate static columns (preserve values if present in df_2024)
for col in static_cols:
    if col in df_2024.columns:
        df_2010[col] = df_2024[col]
    else:
        df_2010[col] = pd.NA

# convert any 2024 columns into 2010equivalents set to 0
for col in df_2024.columns:
    if col.endswith('_2024'):
        new_col = col.replace('_2024', '_2010')
        # avoid overwriting static cols
        if new_col in df_2010.columns:
            continue
        df_2010[new_col] = 0



df_2010.head()

```

```python
import tabula
import pandas as pd
import sys
import os
```

```python
# the headers were messy, this script did not work 
def extract_table_to_excel(pdf_path, excel_path):
    if not os.path.exists(pdf_path):
        print(f"Error: PDF file not found at '{pdf_path}'")
        return

    print(f"Attempting to read tables from: {pdf_path}...")

    try:
        dfs = tabula.read_pdf(pdf_path, pages='1', multiple_tables=True, stream=True)

        if not dfs:
            print("Error: No tables found in PDF.")
            return

        main_table_df = dfs[0]

        print(f"Saving extracted table to: {excel_path}...")
        main_table_df.to_excel(excel_path, index=False, header=True)
        print("File saved")

    except Exception as e:
        print(f"An error occurred: {e}")
        print(f"   PDF Path: {os.path.abspath(pdf_path)}")
        print(f"   Output Path: {os.path.abspath(excel_path)}")
```

```python
pdf_file = "source-files/treb_2025_q3_rental_pg3.pdf"
excel_file = "source-files/rental_data_extraction.xlsx"
    
extract_table_to_excel(pdf_file, excel_file)
```
